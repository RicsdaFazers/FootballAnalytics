```python
#CREATE GOLD DATABASE
spark.sql(
    """
    CREATE DATABASE IF NOT EXISTS UC1_gl LOCATION 'hdfs://hdfs-nn:9000/demo/gold/UC1_gl.db/'
    """
)

#DROP TABLE
spark.sql(
    """
    DROP TABLE IF EXISTS UC1_gl.premier
    """
)

spark.sql(
"""
CREATE EXTERNAL TABLE UC1_gl.premier (
    Div CHAR(10),
    Date CHAR(50),
    HomeTeam CHAR(100),
    AwayTeam CHAR(100),
    Referee CHAR(100),
    FullTimeH_Goals INT,
    FullTimeH_Goals_Avg DOUBLE,
    FullTimeA_Goals INT,
    FullTimeA_Goals_Avg DOUBLE,
    HomeFouls INT,
    HomeFouls_Avg DOUBLE,
    AwayFouls INT,
    AwayFouls_Avg DOUBLE,
    HomeYellowC INT,
    HomeYellowC_Avg DOUBLE,
    AwayYellowC INT,
    AwayYellowC_Avg DOUBLE,
    HomeRedC INT,
    HomeRedC_Avg DOUBLE,
    AwayRedC INT,
    AwayRedC_Avg DOUBLE
    )
    USING DELTA
    LOCATION 'hdfs://hdfs-nn:9000/demo/gold/UC1_gl.db/premier/'
    """
)
```

```python
from pyspark.sql.functions import substring, avg, sum
from pyspark.sql.functions import col, concat_ws, concat, lit, desc, asc

#JOIN THE TABLES
flat_premier_df = premier \
    .withColumn("Date", concat(lit("20"), substring(col("Date"), 7, 8))) \
    .withColumnRenamed("FTHG","FullTimeH_Goals") \
    .withColumnRenamed("FTAG","FullTimeA_Goals") \
    .withColumnRenamed("HF","HomeFouls") \
    .withColumnRenamed("AF","AwayFouls") \
    .withColumnRenamed("HY","HomeYellowC") \
    .withColumnRenamed("AY","AwayYellowC") \
    .withColumnRenamed("HR","HomeRedC") \
    .withColumnRenamed("AR","AwayRedC") \
    .select("Div","Date", "HomeTeam","AwayTeam","Referee","FullTimeH_Goals","FullTimeA_Goals","HomeFouls","AwayFouls","HomeYellowC","AwayYellowC","HomeRedC","AwayRedC") 

#AGGREGATE DATA
premier_df = flat_premier_df \
    .orderBy(asc("Date")) \
    .groupBy("Div", "Date", "HomeTeam", "AwayTeam", "Referee","FullTimeH_Goals","FullTimeA_Goals","HomeFouls","AwayFouls","HomeYellowC","AwayYellowC","HomeRedC","AwayRedC") \
    .agg(
        avg(flat_premier_df.FullTimeH_Goals).alias("FullTimeH_Goals_Avg"),
        avg(flat_premier_df.FullTimeA_Goals).alias("FullTimeA_Goals_Avg"),
        avg(flat_premier_df.HomeFouls).alias("HomeFouls_Avg"),
        avg(flat_premier_df.AwayFouls).alias("AwayFouls_Avg"),
        avg(flat_premier_df.HomeYellowC).alias("HomeYellowC_Avg"),
        avg(flat_premier_df.AwayYellowC).alias("AwayYellowC_Avg"),
        avg(flat_premier_df.HomeRedC).alias("HomeRedC_Avg"),
        avg(flat_premier_df.AwayRedC).alias("AwayRedC_Avg"),
    )

#WRITE TO DELTA TABLE
premier_df \
    .write \
    .format("delta") \
    .mode("overwrite") \
    .save("hdfs://hdfs-nn:9000/demo/gold/UC1_gl.db/premier/")



#CHECK RESULTS
spark.table("UC1_gl.premier").count()
```

```python
# create gold database
spark.sql(
    """
    CREATE DATABASE IF NOT EXISTS gold_table LOCATION 'hdfs://hdfs-nn:9000/demo/gold/gold_table.db/'
    """
)

spark.sql(
    """
    CREATE EXTERNAL TABLE gold_table.premier (
    
        Div STRING,
        Team STRING,
        Date STRING,
        Total_cartoes INT,
        Total_faltas INT,
        Media_faltas DOUBLE,
        Media_cartoes_Amarelos DOUBLE,
        Media_cartoes_Vermelhos DOUBLE,
        Media_golos DOUBLE
        
    )
    USING DELTA
    LOCATION 'hdfs://hdfs-nn:9000/demo/gold/gold_table/premier/'
    """
)
```

```python
from pyspark.sql.functions import substring, avg, sum, round
from pyspark.sql.functions import col, concat_ws, concat, lit, desc, asc

# read data from tables in projeto.db
tabela = spark.read.format("delta").load("hdfs://hdfs-nn:9000/demo/silver/premier_league.db/deltalake_table/")


silver_table = tabela \
    .withColumn("Date", concat(lit("20"), substring(col("Date"), 7, 8)))


silver_table.show()
```

```python
homeTeam_table = silver_table \
    .groupBy("HomeTeam", "Date") \
    .agg(
        sum(silver_table.HY).alias("SHYH"),
        sum(silver_table.HR).alias("SHRH"),
        sum(silver_table.HF).alias("SHFH"),
        avg(silver_table.FTHG).alias("FTHG"),
        avg(silver_table.HF).alias("HFH"),
        avg(silver_table.HY).alias("HYH"),
        avg(silver_table.HR).alias("HRH"),
    ) \
    .withColumnRenamed("HomeTeam", "Team") 

homeTeam_table.show(n=30)
homeTeam_table.count()
```

```python
awayTeam_table = silver_table \
    .groupBy("AwayTeam", "Date", "Div") \
    .agg(
        sum(silver_table.HY).alias("SAYA"),
        sum(silver_table.HR).alias("SARA"),
        sum(silver_table.AF).alias("SAFA"),
        avg(silver_table.FTAG).alias("FTAG"),
        avg(silver_table.AF).alias("AFA"),
        avg(silver_table.AY).alias("AYA"),
        avg(silver_table.AR).alias("ARA"),
    ) \
    .withColumnRenamed("AwayTeam", "Team") 

awayTeam_table.show(n=30)
awayTeam_table.count()
```

```python
#join of the two dataFrames
join_table = homeTeam_table.join(awayTeam_table,['Team', 'Date'], how='left') \
    .withColumnRenamed("Team", "Equipa") \
    .withColumnRenamed("Div", "Liga") \
    .withColumnRenamed("Date", "Data") \
    .orderBy("Date") \
    .withColumn("Media_golos", round(((col("FTAG") + col("FTHG"))/2), 2)) \
    .withColumn("Media_faltas", round(((col("HFH") + col("AFA"))/2), 2)) \
    .withColumn("Media_cartoes_amarelos", round(((col("HYH") + col("AYA"))/2), 2)) \
    .withColumn("Media_cartoes_vermelhos", round(((col("HRH") + col("ARA"))/2), 2)) \
    .withColumn("Total_faltas", round((col("SHFH") + col("SAFA")),2)) \
    .withColumn("Total_cartoes", round((col("SHYH") + col("SAYA") + col("SHRH") + col ("SARA")),2)) \
    .select("Liga","Equipa","Data","Total_cartoes","Total_faltas","Media_faltas","Media_cartoes_Amarelos","Media_cartoes_Vermelhos","Media_golos")


#WRITE TO DELTA TABLE
join_table \
    .write \
    .format("delta") \
    .option("overwriteSchema", "true") \
    .mode("overwrite") \
    .save("hdfs://hdfs-nn:9000/demo/gold/gold_table.db/premier/")



join_table.show()
join_table.count()
```
